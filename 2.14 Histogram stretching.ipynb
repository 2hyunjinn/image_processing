```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
src = cv2.imread('./data/hawkes.bmp', cv2.IMREAD_GRAYSCALE)
img = cv2.imread('./data/hawkes.bmp')

plt.figure(figsize=(7, 7))
plt.axis('off')
plt.imshow(img)
plt.title('Origianl Image')
plt.show()
```

```python
# TODO: Histogram display
plt.hist(src.flatten(), range=[0,256], bins=256);
plt.show()
```

```python
# TODO: Histogram Stretching

# Gmin, Gmax
# min, max 값 , min, max 위치
min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(src)
print('Gmin = ', min_val)
print('Gmax = ', max_val)

#************** 스트레칭하는식 **************
dst = ((src-min_val)*255) / (max_val - min_val) 
dst = cv2.convertScaleAbs(dst) #스트레칭된 영상으로 바꿈
dst2 = cv2.equalizeHist(src) #히스토그램 평활화

cv2.imshow('original',src)
cv2.imshow('stretching', dst) #히스토그램 스트레칭
cv2.imshow('equal', dst2) #히스토그램 평활화
cv2.waitKey()    
cv2.destroyAllWindows()
```
