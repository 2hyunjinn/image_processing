# 1. image read, display

```python
import cv2 #cv2.imshow - 별도의 윈도우에서 출력

imageFile = './data/lena.jpg' #lena 읽음
img = cv2.imread(imageFile) #cv2.IMREAD_COLOR - 뒤에 아무것도 안써져있으면 컬러가 default
print(img.shape)

img2 = cv2.imread(imageFile, 0) # cv2.IMREAD_COLOR - 뒤에 인수가 0이므로 gray로 바꿈
print(img2.shape)

cv2.imshow('Lena color', img) # 채널이 3개임 (512, 512, 3)
cv2.imshow('Lena grayscale', img2) #채널이 1개 (512, 512)

cv2.waitKey()
cv2.destroyAllWindows()
```

# 2. image save

```python
imageFile = './data/lena.jpg'
img = cv2.imread(imageFile)

cv2.imwrite('./out/Lena.bmp', img) #비트맵 저장
cv2.imwrite('./out/Lena.png', img) #png 저장
cv2.imwrite('./out/Lena2.png', img, [cv2.IMWRITE_PNG_COMPRESSION, 9]) #9로 압축
cv2.imwrite('./out/Lena.jpg', img, [cv2.IMWRITE_JPEG_QUALITY, 90])
```

# 3. color image display

```python
from matplotlib import pyplot as plt #plt.imshow - 주피터 안에서 출력

imageFile ='./data/lena.jpg'
imgBGR = cv2.imread(imageFile) #cv2.IMREAD_COLOR
plt.axis('off') #그래프에서 축을 보이지 마라

#컴버트 칼라 => 컬러 모델을 바꿀 수 있음
imgRGB = cv2.cvtColor(imgBGR, cv2.COLOR_BGR2RGB) #BGR -> RGB로 바꾸기위한 코드
plt.imshow(imgRGB)

#imgBGR로 하면 색반전으로 출력
#plt.imshow(imgBGR) # - 푸르스름하게 바뀜

plt.show()
```

# 6.subplot image display

# 여러개의 영상을 한 번에!

```python
import cv2
import matplotlib.pyplot as plt
path = './data/'

imgBGR1 = cv2.imread(path+'lena.jpg')
imgBGR2 = cv2.imread(path+'apple.jpg')
imgBGR3 = cv2.imread(path+'baboon.jpg')
imgBGR4 = cv2.imread(path+'orange.jpg')

#컬러 변환: BGR -> RGB
imgRGB1 = cv2.cvtColor(imgBGR1, cv2.COLOR_BGR2RGB)
imgRGB2 = cv2.cvtColor(imgBGR2, cv2.COLOR_BGR2RGB)
imgRGB3 = cv2.cvtColor(imgBGR3, cv2.COLOR_BGR2RGB)
imgRGB4 = cv2.cvtColor(imgBGR4, cv2.COLOR_BGR2RGB)

#subplots (2,2) - 2행 2열로 이미지 한번에 출력!
fig, ax = plt.subplots(2,2, figsize=(5,5), sharey=True)
#sharey = True? -> y값을 공유하겠다는 뜻임

ax[0][0].axis('off')
ax[0][0].imshow(imgRGB1, aspect = 'auto')

ax[0][1].axis('off')
ax[0][1].imshow(imgRGB2, aspect = 'auto')

ax[1][0].axis('off')
ax[1][0].imshow(imgRGB3, aspect = 'auto')

ax[1][1].axis('off')
ax[1][1].imshow(imgRGB4, aspect = 'auto')

plt.subplots_adjust(left=0, bottom=0,right=1,top=1,wspace=0.05, hspace=0.05)
plt.show()
```
